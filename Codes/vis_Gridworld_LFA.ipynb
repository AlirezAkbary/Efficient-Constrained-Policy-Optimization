```python
import numpy as np
from os import listdir
import matplotlib.pyplot as plt
path = 'GridworldLFA/Results/sMDPO_LSTDSampling/'
folds = [f for f in listdir(path)]
for f in folds:
    try:
        if 'iht_size50' in f:
            og = np.load(path+ f + "/og.npy")
            cv = np.load(path+ f + "/cv.npy")
            fig, axes = plt.subplots(nrows=1, ncols=2)
            axes[0].plot(og)
            axes[1].plot(cv)
            
            plt.title(f+str(np.min(og)))
            fig.tight_layout()
    except:
        continue
```

```python
import numpy as np
from os import listdir
import matplotlib.pyplot as plt
path = 'best_result/Gridworld/sMDPO_LFA/'
folds = [f for f in listdir(path)]
for f in folds:
    if 'og_40' in f:
        sMDPO_40_og = np.load(path+ f)
    elif 'og_50' in f:
        sMDPO_50_og = np.load(path+ f )
    else:
        sMDPO_80_og = np.load(path+ f)
```

```python
alpha = np.arange(len(sMDPO_40_og))
print(len(sMDPO_40_og))
plt.plot([alpha[i] for i in range(0, len(alpha), 100)], [sMDPO_40_og[i] for i in range(0, len(alpha), 100)] , "r", linestyle="-", marker='o', linewidth=2,  label = "d=40 sMDPO")

plt.plot([alpha[i] for i in range(0, len(alpha), 100)], [CBP_40_og[i] for i in range(0, len(CBP_40_og), 10)] , "b", linestyle="-", marker='o', linewidth=2,  label = "d=40 CBP")

plt.plot([alpha[i] for i in range(0, len(alpha), 100)], [GDA_40_og[i] for i in range(0, len(GDA_40_og), 10)] , "g", linestyle="-", marker='o', linewidth=2,  label = "d=40 GDA")


plt.plot([alpha[i] for i in range(0, len(alpha), 100)], [sMDPO_50_og[i] for i in range(0, len(alpha), 100)] , "r", linestyle="-", marker='s', linewidth=2,  label = "d=50 sMDPO")

plt.plot([alpha[i] for i in range(0, len(alpha), 100)], [CBP_50_og[i] for i in range(0, len(CBP_50_og), 10)] , "b", linestyle="-", marker='s', linewidth=2,  label = "d=50 CBP")

plt.plot([alpha[i] for i in range(0, len(alpha), 100)], [GDA_50_og[i] for i in range(0, len(GDA_50_og), 10)] , "g", linestyle="-", marker='s', linewidth=2,  label = "d=50 GDA")


plt.plot([alpha[i] for i in range(0, len(alpha), 100)], [sMDPO_80_og[i] for i in range(0, len(alpha), 100)] , "r", linestyle="-", marker='*', linewidth=2,  label = "d=80 sMDPO")

plt.plot([alpha[i] for i in range(0, len(alpha), 100)], [CBP_80_og[i] for i in range(0, len(CBP_80_og), 10)] , "b", linestyle="-", marker='*', linewidth=2,  label = "d=80 CBP")

plt.plot([alpha[i] for i in range(0, len(alpha), 100)], [GDA_80_og[i] for i in range(0, len(GDA_80_og), 10)] , "g", linestyle="-", marker='*', linewidth=2,  label = "d=80 GDA")

plt.legend(loc="upper right")
plt.xlabel(r'Iteration')
plt.ylabel('Optimality gap')
plt.rcParams.update({'font.size': 20})
plt.tick_params(left = False, bottom=False)
plt.rcParams['figure.figsize'] = 7, 8
plt.savefig('Gridworld-LFA-OG.pdf')
```
